```python
#Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#  Reading the csv

```python
mushroom_train = pd.read_csv('mushroom_train.csv')
mushroom_test = pd.read_csv('mushroom_test.csv')
print(mushroom_train.shape)
print(mushroom_test.shape)
pd.options.display.max_columns = 999
mushroom_train.head()
```

# Data Preprocessing

## Checking for a NaN/null value

```python
print(mushroom_train.loc[:,mushroom_train.isnull().sum()>0].columns) #Checking for a NaN/null value
print(mushroom_test.loc[:,mushroom_test.isnull().sum()>0].columns)
```

## Looking for columns with low variance

Removing columns which have low variance to reduce dimensionality (later on usage of One Hot Encoding dimensionality rises so removing needless features in the beginning)

```python
from sklearn.preprocessing import LabelEncoder
mushroom_train_label= mushroom_train
mushroom_train_label = mushroom_train_label.apply(LabelEncoder().fit_transform)
columns_drop=[] 
#Checking for variance in various features
for col in mushroom_train_label:
    if(mushroom_train_label[col].var()<0.1):
        print(col)
        print(mushroom_train_label[col].var())
#         print(mushroom_train_label[col].value_counts())
        columns_drop.append(col)
# print(columns_drop)# List of columns with low variance
```

## Removing columns with low variance

```python
for col in columns_drop: #dropping columns
    del mushroom_test[col]
    del mushroom_train[col]
```

## Mapping features which have relative values
like in gill-size => broad>narrow

```python
#mapping the features which are relative to each other
class_map  = {'e':0,'p':1} 
bruise_map = {'t':1,'f':0}
gill_spacing = {'c':1,'w':2,'d':3}
gill_size_map = {'b':2,'n':1}
# ring_num_map = {'n':0,'o':1,'t':2}
mushroom_train["class"] = mushroom_train["class"].map(class_map)
mushroom_train["bruises"] = mushroom_train["bruises"].map(bruise_map)
mushroom_train["gill-spacing"] = mushroom_train["gill-spacing"].map(gill_spacing)
mushroom_train["gill-size"] = mushroom_train["gill-size"].map(gill_size_map)
# mushroom_train["ring-number"] = mushroom_train["ring-number"].map(ring_num_map)
mushroom_test["bruises"] = mushroom_test["bruises"].map(bruise_map)
mushroom_test["gill-spacing"] = mushroom_test["gill-spacing"].map(gill_spacing)
mushroom_test["gill-size"] = mushroom_test["gill-size"].map(gill_size_map)
# mushroom_test["ring-number"] = mushroom_test["ring-number"].map(ring_num_map)

print(mushroom_train.iloc[:,1:].shape)
print(mushroom_test.shape)
mushroom_train.head()
```

## One Hot Encoding 

```python
train_objs_num = len(mushroom_train.iloc[:,1:])
print(train_objs_num)
mushroom_train_without_class = mushroom_train.iloc[:,1:]
dataset = pd.concat(objs=[ mushroom_train_without_class,mushroom_test], axis=0)
dataset_preprocessed = pd.get_dummies(dataset)
mushroom_train_enc = dataset_preprocessed[:train_objs_num]
mushroom_test_enc = dataset_preprocessed[train_objs_num:]
print(dataset.shape)
mushroom_train_enc = mushroom_train_enc.join(mushroom_train.iloc[:,0])
print(mushroom_train_enc.shape)
print(mushroom_test_enc.shape)

mushroom_train_enc.head()
```

## Scaling the Dataset
Scaling is done as many algorithms assign weights to features based on the numeric values (to reduce biasing for any particular feature)

```python
#Scaling dataset
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
mushroom_train_scaled = mushroom_train_enc
mushroom_test_scaled = mushroom_test_enc
mushroom_train_scaled.iloc[:,mushroom_train_enc.columns != 'class'] = std.fit_transform(mushroom_train_enc.iloc[:,mushroom_train_enc.columns != 'class'])
mushroom_test_scaled.iloc[:,:] =  std.transform(mushroom_test_enc)
mushroom_train_scaled.head()
```

## Getting a correlation matrix 

```python
#Getting a correlation matrix
corr_matrix = mushroom_train_scaled.iloc[:,mushroom_train_enc.columns != 'class'].corr()
np.fill_diagonal(corr_matrix.values, 0)
print(corr_matrix)
```

## Droppping redundant feature 
So as to reduce dimensionality

```python
#Dropping redundant features
column_num = len(mushroom_train_scaled.iloc[:,mushroom_train_enc.columns != 'class'].columns)
column_name = mushroom_train_scaled.iloc[:,mushroom_train_enc.columns != 'class'].columns
column_drop_enc = []
for i in range(column_num -1):
    for j in range(column_num -1):
        if(i<j and (corr_matrix.iloc[i,j]>0.9)):
            print(column_name[i]+"  "+column_name[j]+" "+str(corr_matrix.iloc[i,j]))
            column_drop_enc.append(column_name[i])
            column_drop_enc.append(column_name[j])
        if(i<j and (corr_matrix.iloc[i,j]<-0.9 )):
            print(column_name[i]+"  "+column_name[j]+" "+str(corr_matrix.iloc[i,j]))
            column_drop_enc.append(column_name[i])
            column_drop_enc.append(column_name[j])
column_drop_enc = list(set(column_drop_enc))
print(column_drop_enc)
print(len(column_drop_enc))
```

```python
mushroom_train_drop = mushroom_train_scaled.drop(column_drop_enc,axis=1)
mushroom_test_drop = mushroom_test_scaled.drop(column_drop_enc,axis=1)
print(mushroom_train_drop.shape)
print(mushroom_test_drop.shape)
mushroom_train_drop.head()
```

# Validation
Spliting data into train and test set to evaluate performance of the model

```python
#Breaking train dataset into test and train for cross validation 
from sklearn.cross_validation import train_test_split
column_num_drop = mushroom_train_drop.shape[1]
X,y=mushroom_train_drop.iloc[:,:column_num_drop-1].values,mushroom_train_drop.iloc[:,column_num_drop-1].values
# column_name = list(mushroom_train_enc.iloc[:,:column_num-1].columns.values)
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3,random_state=0)
print(X_train.shape)
print(y_train.shape)
```

# Selecting Classifier
Based on our analysis tree based classifier should work better

## DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.cross_validation import StratifiedKFold

kfold = StratifiedKFold(y=y_train,n_folds=10,random_state=1)
scores = []

clf = DecisionTreeClassifier()

for k, (train, test) in enumerate(kfold):
    clf.fit(X_train[train], y_train[train])
    score = clf.score(X_train[test], y_train[test])
    scores.append(score)
    print('Fold: %s, Class dist.: %s, Acc: %.3f' % (k+1,np.bincount(y_train[train]), score))

print('CV accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
# print(scores)
print(clf.score(X_test,y_test))
```

## ExtraTreeClassifier

```python
from sklearn.tree import ExtraTreeClassifier 
from sklearn.cross_validation import StratifiedKFold

kfold = StratifiedKFold(y=y_train,n_folds=10,random_state=1)
scores = []

clf = ExtraTreeClassifier()

for k, (train, test) in enumerate(kfold):
    clf.fit(X_train[train], y_train[train])
    score = clf.score(X_train[test], y_train[test])
    scores.append(score)
    print('Fold: %s, Class dist.: %s, Acc: %.3f' % (k+1,np.bincount(y_train[train]), score))

print('CV accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
# print(scores)
print(clf.score(X_test,y_test))
```

## RandomForestClassifier

```python
#based on our analysis using Random forest
from sklearn.ensemble import RandomForestClassifier 
from sklearn.cross_validation import StratifiedKFold

kfold = StratifiedKFold(y=y_train,n_folds=10,random_state=1)
scores = []

clf = RandomForestClassifier(random_state=0)

for k, (train, test) in enumerate(kfold):
    clf.fit(X_train[train], y_train[train])
    score = clf.score(X_train[test], y_train[test])
    scores.append(score)
    print('Fold: %s, Class dist.: %s, Acc: %.3f' % (k+1,np.bincount(y_train[train]), score))

print('CV accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
# print(scores)
print(clf.score(X_test,y_test))
```

## ExtraTreesClassifier

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.cross_validation import StratifiedKFold

kfold = StratifiedKFold(y=y_train,n_folds=10,random_state=1)
scores = []

clf = ExtraTreeClassifier(random_state=0)

for k, (train, test) in enumerate(kfold):
    clf.fit(X_train[train], y_train[train])
    score = clf.score(X_train[test], y_train[test])
    scores.append(score)
    print('Fold: %s, Class dist.: %s, Acc: %.3f' % (k+1,np.bincount(y_train[train]), score))

print('CV accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
# print(scores)
print(clf.score(X_test,y_test))
```
